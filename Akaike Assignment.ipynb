# Importing required libraries:

```python
import requests
from bs4 import BeautifulSoup
from transformers import pipeline
from collections import defaultdict
from gtts import gTTS
import os
from IPython.display import Audio
```

# Creating a function to scrape news articles from a static website:

```python
def scrape_news(company_name):
    url = f"https://www.bbc.co.uk/search?q={company_name}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    articles = []
    for item in soup.find_all('div', class_='ssrcss-1f3bvyz-Stack e1y4nx260'): 
        title = item.find('a').text.strip()
        link = "https://www.bbc.co.uk" + item.find('a')['href']
        summary = item.find('p').text.strip() if item.find('p') else "No summary available."
        articles.append({
            "title": title,
            "link": link,
            "summary": summary
        })
        if len(articles) >= 10:  
            break
    return articles
```

# Creating a function to perform sentiment analysis:

```python
def analyze_sentiment(text):
    sentiment_pipeline = pipeline("sentiment-analysis")
    result = sentiment_pipeline(text)
    return result[0]['label']
```

# Creating a function to perform comparative analysis:

```python
# Function to perform comparative analysis
def comparative_analysis(articles):
    sentiment_count = defaultdict(int)
    topics = defaultdict(int)
    
    for article in articles:
        sentiment = article['sentiment']
        sentiment_count[sentiment] += 1
    
    # Generate insights
    insights = {
        "sentiment_distribution": dict(sentiment_count),
        "overall_sentiment": max(sentiment_count, key=sentiment_count.get)
    }
    return insights
```

# Creating a function to generate hindi TTS:

```python
def generate_hindi_tts(text, filename="output.mp3"):
    tts = gTTS(text=text, lang='hi')
    tts.save(filename)
    return filename
```

```python
# creating a user define variable company_name
company_name = input("Enter Company Name: ")

# Scraping news articles
articles = scrape_news(company_name)

# Performing sentiment analysis on each article
for article in articles:
    title = article['title']
    summary = article['summary']
    sentiment = analyze_sentiment(summary)
    article['sentiment'] = sentiment 
    print(f"Title: {title}")
    print(f"Summary: {summary}")
    print(f"Sentiment: {sentiment}")
    print("-" * 50)

# Performing comparative analysis
insights = comparative_analysis(articles)

# comparative analysis results
print("\nComparative Analysis:")
print(f"Sentiment Distribution: {insights['sentiment_distribution']}")
print(f"Overall Sentiment: {insights['overall_sentiment']}")

# Generating Hindi TTS summary
summary_text = (
    f"{company_name} के बारे में समाचारों का सारांश: "
    f"कुल {len(articles)} लेखों में से {insights['sentiment_distribution'].get('POSITIVE', 0)} सकारात्मक, "
    f"{insights['sentiment_distribution'].get('NEGATIVE', 0)} नकारात्मक, और {insights['sentiment_distribution'].get('NEUTRAL', 0)} तटस्थ हैं।"
)

# TTS output
tts_filename = generate_hindi_tts(summary_text)
print(f"\nHindi TTS summary saved as: {tts_filename}")

Audio(tts_filename)
```
